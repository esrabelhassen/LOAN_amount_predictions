```python
#transforming object to numerical
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for column in data.columns:
  if data[column].dtype=='object':
    data[column]=l.fit_transform(data[column])
data.info()
```

machine learning

```python
from sklearn.model_selection import train_test_split
#selecting feature and target
x=data.drop(' loan_amount',axis=1)#feature
y=data[' loan_amount']#target
#spliting data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#apply linear regression
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
#fitting trainiging and testing set
LR.fit(x_train,y_train)
```

```python
#ploting linear regression
y_pred=LR.predict(x_test)
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 15))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='g')
plt.title ('Linear Regression')
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae_linear = mean_absolute_error(y_test, y_pred)
mse_linear = mean_squared_error(y_test, y_pred)
r2_linear = r2_score(y_test, y_pred)
print(f"MAE: {mae_linear:.2f}")
print(f"MSE: {mse_linear:.2f}")
print(f"R-squared: {r2_linear:.2f}")
```

```python
#from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures()
degree = 4 # degree of the polynomial
0 == PolynomialFeatures(degree = degree)
X_train_poly = pf.fit_transform(x_train)
X_test_poly = pf.transform(x_test)
polynomial_model = LinearRegression()
polynomial_model.fit(X_train_poly , y_train)# fit polynomial regression
y_pred_poly = polynomial_model.predict(X_test_poly)#predection of the loan amount
#metrics for polynomial regression
mae_poly = mean_absolute_error(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
print(f"MAE (Polynomial): {mae_poly:.2f}")
print(f"MSE (Polynomial): {mse_poly:.2f}")
print(f"R-squared (Polynomial): {r2_poly:.2f}")
```

```python
#descision tree
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
tree = tree.DecisionTreeClassifier()
tree.fit(X_train, y_train)   #fitting our model
y_pred=tree.predict(X_test)
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
import seaborn as sns
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Y_test'], colnames=['y_pred'])
sns.heatmap(confusion_matrix, annot=True)#creating confusion matrics
plt.xlabel('Y_PRED')
plt.ylabel('Y_TEST')
```

```python

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,precision_score,f1_score
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion)
accuracy = accuracy_score(y_test, y_pred)#calculating Accuracy
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)#calculating Precision
print("Precision:", precision)
recall = recall_score(y_test, y_pred)# calculating Recall
print("Recall:", recall)
f1=f1_score(y_test,y_pred)
print("f1:",f1)
```
